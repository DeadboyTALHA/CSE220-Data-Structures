```python
#SET A PYTHON
class BTNode:
  def __init__(self, energy):
    self.energy = energy
    self.left = None  # Left child
    self.right = None  # Right child


#Tree Construction
root = BTNode(20)
n1 = BTNode(10)
n2 = BTNode(25)

root.left=n1
root.right=n2


n3 = BTNode(5)
n4 = BTNode(15)
n1.left=n3
n1.right=n4

n5 = BTNode(35)
n2.right=n5

n6 = BTNode(30)
n7 = BTNode(40)
n5.left=n6
n5.right=n7

def calculate_energy(root,target_city_energy):
    total_energy = 1
    current = root

    while current:
        total_energy *= current.energy
        if current.energy == target_city_energy:
            return f"Total energy: {total_energy} tk"
        elif target_city_energy < current.energy:
            current = current.left
        else:
            current = current.right

    return "route does not exist"

# Testing
print(calculate_energy(root, 30))  # Output: Total energy: 525000 tk
print(calculate_energy(root,34))   # Output: route does not exist
```

```python
#SET B PYTHON
class BTNode:
  def __init__(self, tax, city):
    self.tax = tax  # Tax to enter the city
    self.city = city  # City name
    self.left = None  # Left child
    self.right = None  # Right child

#Tree Construction
khulna = BTNode(20, "Khulna")
Ruppur = BTNode(10, "Ruppur")
Bhanga = BTNode(25, "Bhanga")
khulna.right = Bhanga
khulna.left = Ruppur

Pabna = BTNode(5, "Pabna")
Bogra = BTNode(15, "Bogra")
Ruppur.right = Bogra
Ruppur.left = Pabna


Bhulta = BTNode(35, "Bhulta")
Bhanga.right = Bhulta

Rupganj = BTNode(30, "Rupganj")
Sylhet = BTNode(40, "Sylhet")

Bhulta.right = Sylhet
Bhulta.left = Rupganj

def crossing_tax(root, target_city,target_city_tax):
    total_tax = 0
    current = root

    while current:
        total_tax += current.tax
        if current.city == target_city:
            return f"Total Tax: {total_tax-root.tax} tk"
        elif target_city_tax < current.tax:
            current = current.left
        else:
            current = current.right

    return "route does not exist"

# Testing
print(crossing_tax(khulna, "Sylhet",40))  # Output: Total Tax: 100 tk
print(crossing_tax(khulna, "Dhaka",34))   # Output: route does not exist
```
